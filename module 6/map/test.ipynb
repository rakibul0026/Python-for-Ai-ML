```python
import pandas as pd
df=pd.DataFrame({'price':[1,2,3,4],'b':[5,6,7,8]})
df['category']=df['price'].apply(lambda x : x**0.5)
df['A']=df['price'].map(lambda x: x**0.5)
print(df)
                          

```
