List


```python
mark=[2,3,4,5,6,7,8,9]
print(mark[0])
print(len(mark))
print(mark[-1])

```

List Slicing

```python
marks=[2,3,4,5,6,7,8,9]
print(marks[1:4])
print(marks[::2])
print(marks[:2])
print(marks[1:2:6])
print(marks[2:])
print(marks[1:4])

```

List Methods


```python
list=[1,2,3,4,5]
list.append(6)
print(list)

list.insert(0,0)
print(list)

list.remove(3)
print(list)

list.pop()
print(list)

list.sort()
print(list)

list.reverse()
print(list)

```

Tuples in Python


```python
tuple=(1,2,3,4,5)
print(tuple[0])
print(len(tuple))
print(tuple[-1])
print(tuple[1:4])
print(tuple[::2])
print(tuple[:2])
```

Tuples Method


```python
tup=(87,64,33,95,76)
tup.index(33)
print(tup.index(33))
print(tup.count(64))
```
