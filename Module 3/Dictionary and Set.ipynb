Set


```python
num1={1,2,3,4,5}
num1.add(6)
print(num1)

num2={1,2,3,4,5}
num2.update([6,7,8])
print(num2)

num2.remove(3)
print(num2)
```

```python
print(num1.pop())
```

```python
num2.clear()
```

Set Method

```python
a=set.union(num1,num2)
b=set.intersection(num1,num2)
c=set.difference(num1,num2)
d=set.symmetric_difference(num1,num2)


print("Union:",a)
print("Intersection:",b)
print("Difference:",c)
print("Symmetric Difference:",d)
```
