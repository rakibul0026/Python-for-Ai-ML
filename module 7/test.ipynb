```python
#main class
class phone:
    category="Electronics"
    
    #constractor
    def __init__(self,model,battery,camera,battery_percentage):
        self.model=model
        self.battery=battery
        self.camera=camera
        self.battery_percentage=battery_percentage
    #methods
    def charge(self,hour):
     added_charge = hour * 10
     self.battery_percentage += added_charge
      
    def capture(self,photo):
        if self.battery_percentage<=0:
            print("No charge")
        else:
          self.battery_percentage-=photo
          print(f"photo capture in{self.model}")
    class Musicplayer:
        def play_music(self,song_name):
            print(f"playing {song_name}")
    class GPS:
        def show_location(self,city):
            print(f"showing location in {city}")

        
        
       
```

```python
##creating object
apple=phone("iphone",3000,40,2)
apple.capture(1)
print(apple.battery_percentage)

```

```python
#child class
```

```python
class smartphone(phone):
    def __init__(self,model,battery,camera,battery_percentage,ram):
        super().__init__(model,battery,camera,battery_percentage)
        self.ram=ram
```

```python
pro=smartphone("x",4000,50,100,6)
print(pro.model)
print(pro.ram)
```

```python
## Multiple inheritance
```

```python
class smartphone(phone,phone.Musicplayer,phone.GPS):
    def __init__(self,model,battery,camera,battery_percentage,ram):
        super().__init__(model,battery,camera,battery_percentage)
        self.ram=ram
    def browse_internet(self,website):
        print(f"browsing {website}")
    #object of child class
pro=smartphone("x",4000,50,100,6)
pro.play_music("song1")
pro.show_location("city1")
pro.browse_internet("example.com")
```

```python
class Student:
    def __init__(self,name,cgpa):
        self._name=name
        self._cgpa=cgpa
        
    def get_cgpa(self):
        return  self._cgpa
    def set_cgpa(self,new_cgpa):
        if 0.0<=new_cgpa<=4.0:
            self._cgpa=new_cgpa
        else: 
           print("Invalid CGPA value")

#object
student1=Student("rakibul",4)
print(f"Current CGPA:{ student1.get_cgpa()}")
student1.set_cgpa(3.8)

print(f"Current CGPA:{ student1.get_cgpa()}")
print(student1._cgpa)
        
        
```

```python
##Encapsulation

class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.__account_number = account_number
        self.__balance = initial_balance

    def get_account_number(self):
        return self.__account_number

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return True
        return False

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return True
        return False
# Example usage
account = BankAccount("123456789", 1000)
print(f"Account Number: {account.get_account_number()}")
print(f"Initial Balance: {account.get_balance()}")
```

```python
##Polymorphism
```

```python
class Robot:
    def move(self):
        print("The robot moves in a general way")
class LineFollowerRobot(Robot):
        def move(self):
            print("The line follower robot moves by following a line")
class Drone(Robot):
        def move(self):
            print("The drone moves by flying")
#polymorphism Function
def move_robot(robot):
    robot.move()
# Example usage
line_follower = LineFollowerRobot()
drone = Drone()
move_robot(line_follower)  
move_robot(drone)          
```

```python
from abc import ABC,abstractmethod
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass
    def disconnect(self):
        pass
class MySQLDatabase(Database):
    def connect(self):
        print("Connecting to MySQL database...")
    def disconnect(self):
        print("Disconnecting from MySQL database...")
class MongoDBDatabase(Database):
    def connect(self):
        print("Connecting to MongoDB database...")
    def disconnect(self):
        print("Disconnecting from MongoDB database...")
        
#usage
my_db=MySQLDatabase()
my_db.connect()
my_db.disconnect()
```

```python
##super keyword
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name)
        self.breed=breed
    def speak(self):
        print(f"{self.name} barks")
#usage
dog=Dog("Buddy","Labrador")
dog.speak()
print(f"{dog.name} is a {dog.breed}")
```
